```python
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_data(file):
    with open(file, "r") as f:
        data = json.load(f)

    colors = plt.cm.rainbow(np.linspace(0, 1, len(data)))
    for i in range(len(data)):
        x, y = [], []
        for s in data[i]['splits']:
            x.extend(s['split'])
            y += [s['accuracy'] for _ in range(len(s['split']))]
        plt.plot(x, y, color=colors[i], label=round(data[i]['percentage'], 2))

    plt.legend()
    plt.show()
    return data

def plot_percentage(file, p: float = [.33, .20, .10, .05]):
    plt.figure(figsize=(14, 6))
    with open(file, "r") as f:
        data = json.load(f)

    colors = plt.cm.rainbow(np.linspace(0, 1, len(data)))
    for i, k in enumerate(data.keys()):
        if float(k[:4]) not in p:
            continue
    
        x = list(data[k].keys())
        y = [max(data[k][x]) for x in x]
        plt.plot(x, y, label=k, color=colors[i], linewidth=2)
    plt.legend()
    plt.show()
    return data
```

```python
plot_percentage('outputs/removed_layers_numerical.json')
```

```python
def get_fine_grained(data):
    fine_grained = None
    for split in data:
        if split['percentage'] == 1 / 20:
            fine_grained = split
            break
    return fine_grained
```

```python
numerical_data = plot_data("removed_layers_numerical.json")
```

```python
get_fine_grained(numerical_data)
```

```python
grammar_data = plot_data("removed_layers_grammar.json")
get_fine_grained(grammar_data)
```
